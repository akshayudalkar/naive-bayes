# One hot encoding

```python
dftrain = train.iloc[:,0:13]
dftrain = pd.get_dummies(dftrain)
dftrain
```

```python
dftest = test.iloc[:,0:13]
dftest = pd.get_dummies(dftest)
dftest
```

concat train data

```python
datatrain = pd.concat([dftrain, train['Salary']],axis=1)
datatrain
```

```python
datatest = pd.concat([dftest, test['Salary']],axis=1)
datatest
```

Slipt data set

```python
X = datatrain.values[:,0:102]
Y = datatrain.values[:,102]

a = datatest.values[:,0:102]
b = datatest.values[:,102]
```

# Naive Bayes Model

multinomial naive bayes

```python
classifier_mb = MB()
classifier_mb.fit(X,Y)

train_pred_m = classifier_mb.predict(X)
accuracy_train_m = np.mean(train_pred_m==Y)

test_pred_m = classifier_mb.predict(a)
accuracy_test_m = np.mean(test_pred_m==b)

print('Training accuracy is:',accuracy_train_m,'\n','Testing accuracy is:',accuracy_test_m)
```

guassian naive bayes

```python
classifier_gb = GB()
classifier_gb.fit(X,Y) 

train_pred_g = classifier_gb.predict(X)
accuracy_train_g = np.mean(train_pred_g==Y)


test_pred_g = classifier_gb.predict(a)
accuracy_test_g = np.mean(test_pred_g==b)


print('Training accuracy is:',accuracy_train_g,'\n','Testing accuracy is:',accuracy_test_g)
```
